### Data Visualization using Seaborn

#### Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and  informative statistical graphics. 

```python
#sns.__version__ <- tells us the version we have
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')

plt.scatter(x='total_bill',y='tip',data=tips) #using matplotlib
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips) #using seaborn
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips) #using seaborn relplot by default makes a scatter plot
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips, kind='line') 
#using seaborn relplot kind = line makes a line plot
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips , hue='size' , style='sex') 
#using seaborn relplot by default makes a scatter plot and 
#hue adds the legend and changes the colour
#style changes the shape of the marker
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip', data=tips, hue='smoker', size='sex',sizes=(100,500)) 
#using seaborn relplot by default makes a scatter plot
#size takes a tuple or list which changes the size
plt.show()
```

```python
sns.lineplot(x='total_bill',y='tip', data=tips)
plt.show()
```

```python
sns.catplot(x="time",y='tip',data=tips )
#catplot makes by default a strip plot 
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm')
#catplot makes a swarm plot if kind is swarm or we can use sns.swarmplot()
#It is problematic when there are large no. of data points
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', data=tips, hue='smoker',kind='box',dodge=False,fliersize=20,height=6)
#catplot makes a box plot if kind =box or we can use sns.boxplot()
#the dots above the line represent the outliers
#the line below and above the plot shows how many points lie 
#above and below the particular line of pecentile inside the box
#the box area is also called IQR or inter quartile range
#flier size changes the size of the outliers
#height changes the height of the graph
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='boxen')
#catplot makes a boxen plot if kind =boxen or we can use sns.boxenplot()
#the dots above the line represent the outliers
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', data=tips,hue='smoker', kind='violin',height=6)
plt.show()
#the dark area in the plot shows the IQR or inter quartile range
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
plt.show()
```

```python
sns.catplot(x='smoker',data=tips,kind='count')
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker',data=tips,fit_reg=True, height=6)
#lmplot is used to get the best fit line just like in linear regression
plt.show()
```

```python
ans = sns.load_dataset('anscombe')
ans1 = ans[ans['dataset'] == 'I']
sns.lmplot(x='x',y='y',data=ans1,order=2)
plt.show()
```

```python
ans = sns.load_dataset('anscombe')
ans2 = ans[ans['dataset'] == 'II']
sns.lmplot(x='x',y='y',data=ans2,order=2)
#using order we get the best fit
plt.show()
```

```python
ans = sns.load_dataset('anscombe')
ans3 = ans[ans['dataset'] == 'III']
sns.lmplot(x='x',y='y',data=ans3)
plt.show()
```

```python
ans = sns.load_dataset('anscombe')
ans4 = ans[ans['dataset'] == 'IV']
sns.lmplot(x='x',y='y',data=ans4)
plt.show()
```

```python
tips['new_tip'] = tips['tip'] > 5

sns.lmplot(x='total_bill',y='new_tip',data=tips,logistic=True)
#logistic = True tells us that we have two different classes on either side of the line.
plt.show()
```

```python
import pandas as pd

data = pd.read_csv('../ShareRaceByCity.csv')
data
```

```python
data['share_white'] = pd.to_numeric(data.share_white,errors='coerce')
data['share_black'] = pd.to_numeric(data.share_black,errors='coerce')
data['share_native_american'] = pd.to_numeric(data.share_native_american,errors='coerce')
data['share_asian'] = pd.to_numeric(data.share_asian,errors='coerce')
data['share_hispanic'] = pd.to_numeric(data.share_hispanic,errors='coerce')

plt.figure(figsize=(10, 15))

d = data.groupby('Geographic area').mean()
sns.barplot(x='share_white',y=d.index,data=d,dodge=True)
sns.barplot(x='share_black',y=d.index,data=d,dodge=True)
sns.barplot(x='share_native_american',y=d.index,data=d,dodge=True)
sns.barplot(x='share_asian',y=d.index,data=d,dodge=True)
sns.barplot(x='share_hispanic',y=d.index,data=d,dodge=True)

plt.show()
```
